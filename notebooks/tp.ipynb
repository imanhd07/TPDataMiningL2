## Connexion à une base de données MySQL

Dans cette étape nous récupérons des données depuis une base de données relationnelle **MySQL**.

Nous utilisons la bibliothèque **SQLAlchemy** pour créer une connexion entre Python et la base de données.

Actions réalisées :

- création d'une connexion à la base MySQL
- exécution d'une requête SQL
- récupération des résultats dans un DataFrame Pandas
- affichage des étudiants ayant une note finale G3 supérieure à 15


```python
import pandas as pd
from sqlalchemy import create_engine

# Création d'une connexion avec la base de données MySQL
engine = create_engine("mysql+pymysql://root:@localhost/school")

# Requête SQL pour sélectionner les étudiants ayant une note G3 > 15
query = "SELECT * FROM etudiants WHERE G3 > 15"

# stockage du résultat dans un DataFrame Pandas
df = pd.read_sql(query, engine)

# Affichage des 5 premières lignes
print("Apercu des etudiants(G3>15):")
display(df.head())

# Analyse statique des notes(G3)
print("les premieres statistiques descriptives des notes (G3):")
display(df.describe())
```

## Collecte de données par Web Scraping

Le **Web Scraping** consiste à extraire automatiquement des informations à partir de pages web.

Pour cela nous utilisons :

- `requests` pour envoyer des requêtes HTTP
- `BeautifulSoup` pour analyser le contenu HTML
- `pandas` pour organiser les données extraites


Web Scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### Vérification du fichier robots.txt

Avant de réaliser un scraping sur un site web, il est recommandé de consulter le fichier **robots.txt**.

Ce fichier indique quelles parties du site peuvent être explorées par des robots ou scripts automatisés.


```python
#verification du fichier robots.txt du site
url="https://books.toscrape.com/robots.txt"

#Envoi dune requete HTTP pour recuperer le fichier
reponse=requests.get(url)

#Affichage du contenu du fichier
print("Contenu du fichier:")
print(reponse.text)
```

### Extraction des titres des livres

Dans cette étape nous analysons le contenu HTML de la page web.

Les actions réalisées sont :

- téléchargement du contenu HTML de la page
- analyse du HTML avec BeautifulSoup
- recherche des balises `<h3>` contenant les titres des livres
- stockage des titres dans une liste Python


```python
#Url du site
url="https://books.toscrape.com"

#Envoi de la requete vers le site
reponse=requests.get(url)

#Recuperation du contenu HTML de la page
html=reponse.text

#Creation de l`objet BeautifulSoup pour parser le html
soup=BeautifulSoup(html,"html.parser")

#Recherche de toutes les balises <h3> qui contiennent les titres
titres=soup.find_all("h3")
#Creation d`une liste pour stocker les titres
liste_titres=[]

#parcours des resultats 
for t in titres:
    titre=t.text
    liste_titres.append(titre)

#Affichage des titres
print("Liste des titres:")
print(liste_titres)
```

### Transformation des données extraites

Les titres récupérés lors du Web Scraping sont stockés dans une liste Python.

Dans cette étape nous convertissons cette liste en **DataFrame Pandas** afin de faciliter l'analyse et la manipulation des données.


```python
#Creation d`une DataFrame avec les titres
df=pd.DataFrame(liste_titres, columns=["titre"])

#Affichage des 5 premiers lignes
display(df.head())
```

## Acquisition de données via une API

Les **API (Application Programming Interface)** permettent d'accéder à des données en ligne via des requêtes HTTP.

Dans cette étape nous allons :

- envoyer une requête à une API publique
- récupérer les données au format JSON
- convertir ces données en DataFrame Pandas
- afficher les premières lignes du dataset


```python
#url de l`API
api_url="https://jsonplaceholder.typicode.com/posts"

#Envoi de la requete vers l`API
reponse=requests.get(api_url)

#Conversation de la reponse en format JSON
data=reponse.json()

#Creation d`un DataFrame 
df=pd.DataFrame(data)

#Affichage des 5 premiers lignes
display(df.head())
```
