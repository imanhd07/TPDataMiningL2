# Web Scraping des données

Dans cette étape, nous récupérons des données depuis une page web.

La bibliothèque **Requests** est utilisée pour envoyer une requête HTTP, et **BeautifulSoup** pour analyser le contenu HTML.

Actions réalisées dans la cellule suivante :

- envoi d’une requête HTTP avec `requests.get()`
- vérification du statut de la réponse
- analyse du HTML avec `BeautifulSoup`
- extraction des noms de pays
- affichage d’un échantillon de données

```python
import requests
from bs4 import BeautifulSoup
import time
#url de test dedies a l`apprentissage
url="https://www.scrapethissite.com/pages/simple"
response=requests.get(url)

if response.status_code==200:


    soup = BeautifulSoup(response.text, "html.parser")
    # Extraction des noms de pays (Balise h3 avec classe specifique)
    countries = soup.find_all("h3", class_="country-name")
    print("---␣Echantillon␣des␣donnees␣collectees␣---")
    for country in countries[:5]:
          print(f"Pays␣:␣{country.text.strip()}")
else:

           print(f"Echec␣de␣l’acquisition.␣Code␣d’erreur␣:␣{response.status_code}")
```

# Ingestion des données JSON

Dans cette étape, nous manipulons des données au format JSON.

La bibliothèque **json** permet de charger les données, et **Pandas** de les transformer en DataFrame.

Actions réalisées dans la cellule suivante :

- chargement des données JSON avec `json.loads()`
- conversion en structure Python (liste de dictionnaires)
- transformation en DataFrame avec `pd.DataFrame()`

```python
import pandas as pd
import json

# [span_10](start_span)Flux de données simulant une API[span_10](end_span)
json_raw = """
[
    {"id": 1, "ville": "Nouakchott", "temp": 32.5, "humidite": 45},
    {"id": 2, "ville": "Nouadhibou", "temp": 28.0, "humidite": 60},
    {"id": 3, "ville": "Atar", "temp": 38.2, "humidite": 20},
    {"id": 4, "ville": "Zouerate", "temp": 35.5, "humidite": 15}
]
"""

# 1. [span_11](start_span)Chargement et conversion[span_11](end_span)
data = json.loads(json_raw)
df = pd.DataFrame(data)

# 2. [span_12](start_span)Calculs statistiques (TAF)[span_12](end_span)
moyenne_temp = df['temp'].mean()
max_humidite = df['humidite'].max()

print(f"Température moyenne : {moyenne_temp}")
print(f"Humidité maximale : {max_humidite}")

# 3. [span_13](start_span)Export en CSV avec séparateur ';'[span_13](end_span)
df.to_csv('acquisition_resultat.csv', sep=';', index=False)
```

# Extraction depuis une base de données MySQL

Dans cette étape, nous récupérons des données depuis une base de données relationnelle.

La bibliothèque **SQLAlchemy** permet de se connecter à la base, et **Pandas** d’exécuter la requête.

Actions réalisées dans la cellule suivante :

- configuration des paramètres de connexion
- création du moteur avec `create_engine()`
- écriture d’une requête SQL
- exécution avec `pd.read_sql()`
- affichage des résultats

```python
import pandas as pd
from sqlalchemy import create_engine

# Paramètres de connexion à la base de données
user = "root"
password = ""
host = "localhost"
db_name = "cinema_db"

# Création du moteur de connexion
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}/{db_name}")

# Requête SQL pour récupérer les films Sci-Fi
query = "SELECT movie_id, title, genres FROM movies WHERE genres LIKE '%%Sci-Fi%%'"

# Exécution de la requête et chargement dans un DataFrame
try:
    df_movies = pd.read_sql(query, engine)
    print("Donnees recuperees avec succes !")
    print(df_movies.head())
except Exception as e:
    print(f"Erreur lors de la connexion : {e}")
```
